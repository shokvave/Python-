Array operation (Mathematical manipulations)

```python
a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])

c = a - b
print(c)

d = a*b
print(d)
```

```python
#farenheit to celcious value
farenheit = np.array([0, -10, -15, 0])
# conversion formula = ((F - 32 ) * 5/9 = c)
celcious = (farenheit - 32) * 5/9 
celcious
```

```python
#boolean array
celcious > -20 

```

```python
#Beside  elementwise manipulation numpy also supports matrix manipulations
A = np.array([[1, 1] , [0, 1]])
B = np.array([[2, 0] , [3, 4]])
#if we want matrix wise manipulation we have to use "@" instead of "*" which we use for element wise product
print(A@B)


```

when manipulating array of two diffrent types the type of resulting array will correspond to the more general of two types , this is called Upcasting 


```python
array1 = np.array([[1, 2, 3] , [4, 5, 6]])
array2 = np.array([[7.1, 8.2, 9.3] , [10.4, 11.5, 12.6]])

array3 = array1 + array2
print(array3)
print(array3.dtype)
```

```python
print(array3.sum())
print(array3.max())
print(array3.min())
print(array3.mean())
```

```python
mask = np.full(array.shape, 188)
mask
```
