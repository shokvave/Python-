```python
def quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
    
    
```

```python
quality(16.99, 1.01)
```

```python
df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1)
 
```

```python
df['Quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1)
```

# they both do the same this but look at the 'vectorsized' it imporoves computational power and is less complicated


```python
df['Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])#vectorised method call
```
